```python
from google.colab import files
uploaded=files.upload()

#import pandas as pd
#vcl=pd.read_csv("Electric_Vechicle_Population_Data.csv")
#print(vcl)

import os
os.listdir()
#os.remove("Electric_Vehicle_Population_Data (1).csv")
import pandas as pd

vcl = pd.read_csv("Electric_Vehicle_Population_Data.csv")
print(vcl.head())
vcl.info()
vcl.isna().sum()

ele=vcl['Electric Range'].fillna(vcl['Electric Range'].median(), inplace=True)
print(ele)

loc=location_cols = [
    'County', 'City', 'Postal Code',
    'Electric Utility', '2020 Census Tract'
]

vcl[location_cols] = vcl[location_cols].fillna("Unknown")

print(loc)

kl=vcl['Legislative District'].fillna("Not Assigned", inplace=True)
print(kl)


vcl['Vehicle Location'].fillna("POINT (0 0)", inplace=True)
vcl[['Longitude', 'Latitude']] = (
    vcl['Vehicle Location']
    .str.extract(r'POINT \(([-\d.]+) ([-\d.]+)\)')
    .astype(float)
)
vcl.isna().sum()
vcl.info()

top=vcl['Make'].value_counts().head()
print("top ev makes and models are:",top)


country=vcl['Model Year'].value_counts().sort_index().head()
print("top country are:",country)

adp=vcl['County'].value_counts().head()
print("ev adoption changer over different year re:",adp)

avg=vcl['Electric Range'].describe()
print("average rate of ev are: ",avg)


ca=vcl.groupby('Electric Vehicle Type')['Electric Range'].mean()
print("perecentage of ev are avilable for CAFV are:",ca)


model=vcl.groupby('Model')['Electric Range'] \
  .mean() \
  .sort_values(ascending=False) \
  .head(10)
print("electric ranges accrros different model are:",model)
vcl.columns


cnt=vcl['County'].value_counts().head(10)
print("country wise trends are:",cnt)

mk=vcl.groupby('Make')['Electric Range'].mean().sort_values(ascending=False).head()
print("electric ranges accross d/f make",mk)

import matplotlib.pyplot as plt

year_counts = vcl['Model Year'].value_counts().sort_index()

plt.figure()
year_counts.plot(kind='line')
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.title("EV Adoption Trend Over Years")
plt.show()


top_makes = vcl['Make'].value_counts().head(10)

plt.figure()
top_makes.plot(kind='bar')
plt.xlabel("EV Make")
plt.ylabel("Number of Vehicles")
plt.title("Top 10 EV Manufacturers")
plt.show()

top_counties = vcl['County'].value_counts().head(10)

plt.figure()
top_counties.plot(kind='bar')
plt.xlabel("County")
plt.ylabel("Number of EVs")
plt.title("Top 10 Counties by EV Registrations")
plt.show()


plt.figure()
plt.hist(vcl['Electric Range'], bins=30)
plt.xlabel("Electric Range (miles)")
plt.ylabel("Frequency")
plt.title("Distribution of Electric Range")
plt.show()

avg_range_type = vcl.groupby('Electric Vehicle Type')['Electric Range'].mean()

plt.figure()
avg_range_type.plot(kind='bar')
plt.xlabel("Vehicle Type")
plt.ylabel("Average Electric Range")
plt.title("Average Electric Range by EV Type")
plt.show()

cafv_counts = vcl['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()

plt.figure()
cafv_counts.plot(kind='bar')
plt.xlabel("CAFV Eligibility Status")
plt.ylabel("Number of Vehicles")
plt.title("CAFV Eligibility Distribution")
plt.show()




```
